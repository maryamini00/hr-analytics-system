```python
# birth date
dates = pd.date_range('1960-01-01', '2007-01-01', freq=' D')
birth_date = np.random.choice(dates, size= 10, replace= True)
d3 = pd.DataFrame(birth_date)
# marital status
today = pd.Timestamp.today()

marital_status = []
for a in birth_date:
    a = (today - a).days // 365
    if a <= 22:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.95, 0.05]))
    elif a <= 27:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.6, 0.4]))
    elif a <= 35:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.3, 0.7]))
    elif a <= 50:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.15, 0.85]))
    else:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.1, 0.9]))
        
d3["marital_status"] = marital_status
d3.rename(columns= { 0: "birth_date"}, inplace=True)
d3
```

```python
# Job group
job_group_kind = ["وکلا و مشاوران حقوقی", "پشتیبانی حقوقی", "ستادی", "مدیران"]
d3["job_group"] = np.random.choice(job_group_kind, p= [0.45, 0.2, 0.2, 0.15])


# field of study
field = []
for a in d3["job_group"]:
    if a == "وکلا و مشاوران حقوقی":
        field.append(np.random.choice(["حقوق خصوصی", "حقوق عمومی", "حقوق تجارت", "حقوق بین‌الملل", "حقوق کیفری"]))
    elif a == "پشتیبانی حقوقی":
        field.append(np.random.choice(["حقوق", "مدیریت", "علوم اداری", "حقوق خصوصی"]))
    elif a == "ستادی":
        field.append(np.random.choice(["مدیریت", "حسابداری", "مهندسی کامپیوتر", "مدیریت منابع انسانی", "روابط عمومی"]))
    else:
        field.append(np.random.choice(["حقوق", "MBA", "مدیریت", "حقوق تجارت"]))
        
d3["field_of_study"] = field


# education level
education_level = []
for i in range(len(d3)):
    b = d3["job_group"].iloc[i]
    a = (today - d3["birth_date"].iloc[i]).days // 365
    if a >= 35:
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.65, 0.05]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.2, 0.5, 0.3]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.25, 0.45, 0.3]))
        else:
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.4, 0.5, 0.1]))
    elif (a < 35) & (a >= 28):
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.6, 0.1]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.7]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.15, 0.35, 0.5]))
        else:
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.5, 0.2]))
    elif (a < 28) & (a >= 24):
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.05, 0.65, 0.3]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.7]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.35, 0.65]))
        else:
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.05, 0.7, 0.25]))
    else:
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.1, 0.9]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.05, 0.95]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.21, 0.79]))
        else:
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.2, 0.8]))

d3["education_level"] = education_level
# university
```
