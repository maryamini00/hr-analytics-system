```python
# manager id assign
hierarchy = {
    # dept_name : (positions_of_employees_in_this_dept, positions_of_their_managers)
    "فناوری اطلاعات":           (["کارشناس IT", "کارشناس IT ارشد"],
                                ["مدیر IT"]),
    "مالی و حسابداری":          (["کارشناس مالی", "کارشناس مالی ارشد"],
                                ["مدیر مالی"]),
    "منابع انسانی":             (["کارشناس منابع انسانی", "کارشناس منابع انسانی ارشد"],
                                ["مدیر منابع انسانی"]),
    "اداری و پشتیبانی":         (["کارشناس اداری", "کارشناس اداری ارشد","کارشناس اداری حقوقی", "کارشناس روابط عمومی","کارشناس روابط عمومی ارشد"],
                                ["مدیر اداری"]),
    "دپارتمان اسناد و مدارک":   (["کارشناس اسناد", "کارشناس اسناد ارشد"],
                                ["مشاور اسناد ارشد"]),
    "دپارتمان تحقیقات حقوقی":   (["تحلیلگر حقوقی ارشد"],
                                ["مدیر پژوهش حقوقی"]),
    "دپارتمان حقوقی قراردادها": (["وکیل پایه", "وکیل ارشد", "کارشناس قرارداد ارشد","مشاور حقوقی"],
                                ["تحلیلگر قرارداد ارشد"]),
    "مشاوره حقوقی شرکتی":       (["وکیل پایه", "وکیل ارشد", "مشاور حقوقی","مشاور حقوقی ارشد"],
                                ["وکیل متخصص"]),
    "مدیریت ارشد":              (["مدیر IT", "مدیر مالی", "مدیر منابع انسانی", "مدیر اداری","مشاور اسناد ارشد", "مدیر پژوهش حقوقی","تحلیلگر قرارداد ارشد", "وکیل متخصص","مدیر واحد حقوقی", "مدیر عملیات", "مدیر روابط عمومی"],
                                ["مدیر ارشد عملیات", "معاون مدیریت"]),
}

active = d["employment_status"] == "فعال"
d["manager_id"] = pd.NA
d["manager_id"] = d["manager_id"].astype("Int64")

ceo_idx = d[d["position"] == "CEO"].index[0]

for dept, (emp_pos, mgr_pos) in hierarchy.items():
    mgr_idx = d[active & d["position"].isin(mgr_pos) & (d["department"] == dept)].index.tolist()
    emp_idx = d[active & d["position"].isin(emp_pos) & (d["department"] == dept)
                & ~d.index.isin(mgr_idx)].index.tolist()

    if not emp_idx:
        continue
    if not mgr_idx:
        raise ValueError(f"'{dept}' کارمند دارد ولی مدیر فعال ندارد.")
    if len(mgr_idx) > len(emp_idx):
        raise ValueError(f"'{dept}': مدیران ({len(mgr_idx)}) از کارکنان ({len(emp_idx)}) بیشترند.")

    for i, eidx in enumerate(emp_idx):
        d.at[eidx, "manager_id"] = mgr_idx[i % len(mgr_idx)]

senior_idx = d[active & d["position"].isin(["مدیر ارشد عملیات", "معاون مدیریت"])].index.tolist()
for idx in senior_idx:
    d.at[idx, "manager_id"] = ceo_idx

d
```
