### import libraries

```python
import numpy as np, time
import pandas as pd 
from faker import Faker as fk 
import matplotlib.pyplot as plt
fake = fk("fa_IR")
```

### Basic personal information

```python
# national_code = 03 + 7/8 + 7 ranom number
d = pd.DataFrame({
    "n1": "03",
    "n2": np.random.choice([7, 8], size = 85).astype(str),
    "n3": np.random.randint(1000000, 10000000, size = 85).astype(str)
})

d["national_code"] = d["n1"] + d["n2"] + d["n3"]
d = d.drop(["n1", "n2", "n3"], axis=1)

# phone number = 09 + 9 random number
d["n1"] = "09"
d["n2"] = np.random.choice(["12", "19", "10"], size = 85)
d["n3"] = np.random.randint(1000000, 10000000, size = 85).astype(str)
d["phone_number"] = d["n1"] + d["n2"] + d["n3"]
d = d.drop(["n1", "n2", "n3"], axis=1)

# emergency contact   phone + name
d["n1"] = "09"
d["n2"] = np.random.choice(["12", "19", "10"], size = 85)
d["n3"] = np.random.randint(1000000, 10000000, size = 85).astype(str)
d_ecn = [{
    "emergency_contact_name": fake.name()
}
    for _ in range(85)]
d["emergency_contact_phone"] = d["n1"] + d["n2"] + d["n3"]
d = d.drop(["n1", "n2", "n3"], axis=1)
    
d_1 = pd.DataFrame(d_ecn)
d = pd.concat([d, d_1], axis=1)
```

```python
# first name
# last name
# father name
# gender
d_f = [{
    "first_name": fake.first_name_female(),
    "last_name": fake.last_name(),
    "gender": "خانم",
    "father_name": fake.first_name_male(),
    "email": fake.company_email()
}
    for _ in range(30)]
d_m = [{
    "first_name": fake.first_name_male(),
    "last_name": fake.last_name(),
    "gender": "آقا",
    "father_name": fake.first_name_male(),
    "email": fake.company_email()
}
    for _ in range(55)]
d_s = d_f + d_m
d_s = pd.DataFrame(d_s)
d = pd.concat([d, d_s], axis=1)
```

```python
# birth date
dates = pd.date_range('1960-01-01', '2008-01-01', freq=' D')
birth_date = np.random.choice(dates, size= 85, replace= True)
d_s = pd.DataFrame(birth_date)
d = pd.concat([d, d_s], axis=1)

# marital status
today = pd.Timestamp.today()
marital_status = []
for a in birth_date:
    a = (today - a).days // 365
    if a <= 22:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.95, 0.05]))
    elif a <= 27:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.6, 0.4]))
    elif a <= 35:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.3, 0.7]))
    elif a <= 50:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.15, 0.85]))
    else:
        marital_status.append(np.random.choice(["مجرد", "متاهل"], p= [0.1, 0.9]))
        
marital_status = pd.DataFrame(marital_status)
d["marital_status"] = marital_status
d.rename(columns= { 0: "birth_date"}, inplace=True)
```

```python
# address
d_s = [{
    "address": fake.street_address(),
}
    for _ in range(85)]
d["address"] = pd.DataFrame(d_s)
```

```python
# Job group
job_group_kind = ["وکلا و مشاوران حقوقی", "پشتیبانی حقوقی", "ستادی", "مدیران"]
d["job_group"] = np.random.choice(job_group_kind, size= 85, p= [0.45, 0.2, 0.2, 0.15])


# field of study
field = []
for a in d["job_group"]:
    if a == "وکلا و مشاوران حقوقی":
        field.append(np.random.choice(["حقوق خصوصی", "حقوق عمومی", "حقوق تجارت", "حقوق بین‌الملل", "حقوق کیفری"]))
    elif a == "پشتیبانی حقوقی":
        field.append(np.random.choice(["حقوق", "مدیریت", "علوم اداری", "حقوق خصوصی"]))
    elif a == "ستادی":
        field.append(np.random.choice(["مدیریت", "حسابداری", "مهندسی کامپیوتر", "مدیریت منابع انسانی", "روابط عمومی"]))
    else:
        field.append(np.random.choice(["حقوق", "MBA", "مدیریت", "حقوق تجارت"]))
        
d["field_of_study"] = field


# education level
education_level = []
for i in range(len(d)):
    b = d["job_group"].iloc[i]
    a = (today - d["birth_date"].iloc[i]).days // 365
    if a >= 35:
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.65, 0.05]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.2, 0.5, 0.3]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.25, 0.45, 0.3]))
        else:
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.4, 0.5, 0.1]))
    elif (a < 35) & (a >= 27):
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.6, 0.1]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.7]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.15, 0.35, 0.5]))
        else:
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.5, 0.2]))
    elif (a < 27) & (a >= 24):
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.0, 0.7, 0.3]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.3, 0.7]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.35, 0.65]))
        else:
            education_level.append(np.random.choice(["دکتری", "کارشناسی ارشد", "کارشناسی"], p= [0.0, 0.7, 0.3]))
    else:
        if b == "وکلا و مشاوران حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.1, 0.9]))
        elif b == "پشتیبانی حقوقی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.05, 0.95]))
        elif b == "ستادی":
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.21, 0.79]))
        else:
            education_level.append(np.random.choice(["کارشناسی ارشد", "کارشناسی"], p= [0.2, 0.8]))

d["education_level"] = education_level

# university
uni_kind = ["sorbonne", "UCL", "امیرکبیر", "شریف", "تهران", "manchester", "شهید بهشتی", "علامه", "شیراز", "الزهرا", "آزاد"]
university = []
for i in range(len(d)):
    if d["education_level"].iloc[i] == "دکتری":
        university.append(np.random.choice(uni_kind)) 
    elif d["education_level"].iloc[i] == "کارشناسی ارشد":
        university.append(np.random.choice(uni_kind))
    else:
        university.append(np.random.choice(uni_kind))
d["university"] = university
```

```python
# position
position = []
for i in range(len(d)):
    if d["job_group"].iloc[i] == "وکلا و مشاوران حقوقی":
        if d["education_level"].iloc[i] == "کارشناسی":
            position.append("وکیل پایه")
        elif d["education_level"].iloc[i] == "کارشناسی ارشد":
            position.append(np.random.choice(["مشاور حقوقی", "کارشناس قرارداد ارشد"], p= [0.4, 0.6]))
        else:
            position.append(np.random.choice(["وکیل ارشد", "وکیل متخصص", "مشاور حقوقی ارشد"], p= [0.45, 0.3, 0.25]))
            
    elif d["job_group"].iloc[i] == "پشتیبانی حقوقی":
        if d["education_level"].iloc[i] == "کارشناسی":
            position.append(np.random.choice(["کارشناس اداری حقوقی", "کارشناس اسناد"], p= [0.5, 0.5]))
        elif d["education_level"].iloc[i] == "کارشناسی ارشد":
            position.append(np.random.choice(["کارشناس اسناد ارشد", "تحلیلگر قرارداد ارشد"], p= [0.4, 0.6]))
        else:
            position.append(np.random.choice(["مدیر پژوهش حقوقی", "تحلیلگر حقوقی ارشد", "مشاور اسناد ارشد"], p= [0.4, 0.35, 0.25]))
            
    elif d["job_group"].iloc[i] == "ستادی":
        if d["education_level"].iloc[i] == "کارشناسی":
            if d["field_of_study"].iloc[i] == "مدیریت منابع انسانی":
                position.append("کارشناس منابع انسانی")
            elif d["field_of_study"].iloc[i] == "روابط عمومی":
                position.append("کارشناس روابط عمومی")
            elif d["field_of_study"].iloc[i] == "مهندسی کامپیوتر":
                position.append("کارشناس IT")
            elif d["field_of_study"].iloc[i] == "حسابداری":
                position.append("کارشناس مالی")
            else:
                position.append("کارشناس اداری")
        else:
            if d["field_of_study"].iloc[i] == "مدیریت منابع انسانی":
                position.append("کارشناس منابع انسانی ارشد")
            elif d["field_of_study"].iloc[i] == "روابط عمومی":
                position.append("کارشناس روابط عمومی ارشد")
            elif d["field_of_study"].iloc[i] == "مهندسی کامپیوتر":
                position.append("کارشناس IT ارشد")
            elif d["field_of_study"].iloc[i] == "حسابداری":
                position.append("کارشناس مالی ارشد")
            else:
                position.append("کارشناس اداری ارشد")
    else:
        if d["education_level"].iloc[i] == "کارشناسی":
            position.append(np.random.choice(["مدیر IT", "مدیر مالی", "مدیر اداری"], p=[0.3, 0.35, 0.35]))
        elif d["education_level"].iloc[i] == "کارشناسی ارشد":
            position.append(np.random.choice(["مدیر منابع انسانی", "مدیر عملیات", "مدیر روابط عمومی"], p=[0.35, 0.35, 0.3]))
        else:
            position.append(np.random.choice(["مدیر واحد حقوقی", "مدیر ارشد عملیات", "معاون مدیریت"], p=[0.4, 0.35, 0.25]))
d["position"] = position
```

```python
# hire date
hire_date = []
for i in range(len(d)):
    if d["education_level"].iloc[i] == "دکتری":
        min_year = pd.to_timedelta(9900, unit='D')
    elif d["education_level"].iloc[i] == "کارشناسی ارشد":
        min_year = pd.to_timedelta(8800, unit='D')
    else:
        min_year = pd.to_timedelta(8100, unit='D')
        
    min_date = d["birth_date"].iloc[i] + min_year
    
    end_date = pd.Timestamp('2026-01-01')
    if min_date >= end_date:
        min_date = end_date - pd.to_timedelta(365, unit='D')

    hire_date_s = np.random.choice(pd.date_range(min_date, end_date, freq='D'))
    hire_date.append(hire_date_s)

d["hire_date"] = hire_date
d["hire_date"].notna().count()
```
